```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```python
PCListurl = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
PCCoordurl = "https://cocl.us/Geospatial_data"
```

```python
def getHTMLContent(PCListurl):
    html=urlopen(PCListurl)
    soup=BeautifulSoup(html, 'html.parser')
    return soup
```

```python
content = getHTMLContent(PCListurl)
tables = content.find_all('table')
for table in tables:
    print(table.prettify())
```

```python
req=requests.get(PCListurl)
soup2=BeautifulSoup(req.content, 'lxml')
table2=soup2.find_all('table')[0]
df=pd.read_html(str(table2))
neighborhoodDF=pd.DataFrame(df[0])
neighborhoodDF.head()
```

```python
# drop rows with "NaN" in "Neighborhood" column, as "NaN" neighborhoods do not have Boroughs
# remove/replace punctuation

neighborhoodDF.dropna(subset=["Neighborhood"], axis=0, inplace=True)
neighborhoodDF["Neighborhood"]=neighborhoodDF["Neighborhood"].str.replace(" /", ",")

# reset index, because we dropped many rows
neighborhoodDF.reset_index(drop=True, inplace=True)

neighborhoodDF
```

```python
# Number of rows
neighborhoodDF.shape
print(f"Toronto has {neighborhoodDF.shape[0]} assigned postal codes.")
```

**Combine dataframe with Lat/Long dataframe provided**

```python
PCDF=pd.read_csv(PCCoordurl)
PCDF.head()
```

```python
CombDF=pd.merge(left=neighborhoodDF, right=PCDF, how='inner', left_on='Postal code', right_on='Postal Code').drop('Postal Code', axis=1)
CombDF
```
